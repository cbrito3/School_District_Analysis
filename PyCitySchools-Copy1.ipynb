```python
# Files to load
school_data_to_load = os.path.join("Resources", "schools_complete.csv")
student_data_to_load = os.path.join("Resources", "students_complete.csv")
```

```python
# Read the school data file and store it in a Pandas DataFrame.
school_data_df = pd.read_csv(school_data_to_load)
school_data_df
```

```python
# Read the student data file and store it in a Pandas DataFrame.
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
# 4.5.1 Find Missing Values 
#Determine if there are any missing values in the school data.
school_data_df.count()
```

```python
# Determine if there are any missing values in the student data.
student_data_df.count()
```

```python
#The isnull() Method
# Determine if there are any missing values in the school data.
school_data_df.isnull()

#When we execute this code, we see that every row that is not empty is given the Boolean value "False," which tells us that there are no missing values.
```

```python
# Determine if there are any missing values in the student data.
student_data_df.isnull()
```

```python
#To get the total number of empty rows, or rows that are "True," 
#we can use the Pandas sum() method after the isnull() method, like this

# Determine if there are any missing values in the student data.
student_data_df.isnull().sum()
```

```python
#The notnull() Method
#Another method that we can use to find missing values is the notnull() method. When you apply the notnull() method 
#to a column, Series, or a DataFrame, a Boolean will be returned: "True" for the row or rows that are not empty, 
#or "False" for the row or rows that are empty. This method returns the opposite output of the isnull() method.
```

```python
#Let's use the notnull() method on the school_data_df DataFrame. 
# Determine if there are not any missing values in the school data.
school_data_df.notnull()

#When we run this code, the output returns a copy of our school_data_df DataFrame, 
#where all the rows that do not have any missing values are labeled "True."
```

```python
# Determine if there are not any missing values in the student data.
student_data_df.notnull().sum()
```
